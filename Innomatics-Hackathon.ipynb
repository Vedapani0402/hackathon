```python
from math import radians, cos, sin, asin, sqrt

def haversine(lon1, lat1, lon2, lat2):
    """
    Calculate the great circle distance in kilometers between two points 
    on the earth (specified in decimal degrees)
    """
    # convert decimal degrees to radians 
    lon1, lat1, lon2, lat2 = map(radians, [lon1, lat1, lon2, lat2])

    # haversine formula 
    dlon = lon2 - lon1 
    dlat = lat2 - lat1 
    a = sin(dlat/2)**2 + cos(lat1) * cos(lat2) * sin(dlon/2)**2
    c = 2 * asin(sqrt(a)) 
    r = 6371 # Radius of earth in kilometers. Use 3956 for miles. Determines return value units.
    return c * r
```

```python
data1['hdistance']=data1.apply(lambda x: haversine(x['pickup_longitude'],x['pickup_latitude'],x['dropoff_longitude'],x['dropoff_latitude']),axis=1)
```

### Train-test-Split

```python
from sklearn.model_selection import train_test_split
x=data1[['passenger_count', 'hdistance', 'week_day']]
y=data1['fare_amount']
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=100)
```

#### Pre-processing (Scaling and Encodiing)

```python
from sklearn.preprocessing import OneHotEncoder, LabelEncoder , StandardScaler
onc=OneHotEncoder(drop='first',handle_unknown='ignore')
onc.fit(x_train[['week_day']])
x_train_cat=pd.DataFrame(onc.transform(x_train[['week_day']]).toarray(), columns=onc.get_feature_names())
x_test_cat=pd.DataFrame(onc.transform(x_test[['week_day']]).toarray(), columns=onc.get_feature_names())
```

```python
ncols=['passenger_count', 'hdistance']
scaler = StandardScaler()
scaler.fit(x_train[ncols])
x_train_std = pd.DataFrame(scaler.transform(x_train[ncols]),columns=x_train[ncols].columns)
x_test_std = pd.DataFrame(scaler.transform(x_test[ncols]),columns=x_train[ncols].columns)
```

```python
x_train_con=pd.concat((x_train_cat,x_train_std),axis=1)
x_test_con=pd.concat((x_test_cat,x_test_std),axis=1)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
```

### Model Training 

#### Linear Regression

```python
linreg = LinearRegression()
model_l=linreg.fit(x_train_con, y_train)
```

```python
r_sqr = model_l.score(x_test_con, y_test)
n= x_test.shape[1]
p = len(y_test)
adjusted_r = 1 - (1 - r_sqr) * (p - 1) / (n - p-  1)
adjusted_r
```

#### Random Forest Regressor

```python
rfr = RandomForestRegressor()
model_r=rfr.fit(x_train_con, y_train)
```

```python
r_sqr = model_r.score(x_test_con, y_test)
n= x_test.shape[1]
p = len(y_test)
adjusted_r = 1 - (1 - r_sqr) * (p - 1) / (n - p-  1)
adjusted_r
```
